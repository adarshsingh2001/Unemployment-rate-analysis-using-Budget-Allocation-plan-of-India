# Analysis of Unemployment Rate Change according to Budget Allocation

Import all the Required Packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Importing India's Budget Allocation Dataset 2020 and storing it in a panadas dataframe

```python
budget_2020 = pd.read_csv("India_budget_2020.csv")
budget_2020
```

To know the Statistical property of our dataset

```python
budget_2020.describe()
```

```python
budget_2020.info()
```

Dropping all the empty / Null values present in our dataset

```python
budget_2020.dropna()
```

Dropping the 'Total' row because we don't need them in our analysis

```python
budget_2020 = budget_2020.drop(56,axis=0)
budget_2020
```

Since not all the department are equally important for our analysis, we categorize 12 most important department and store other dept in 'Others' section

```python
imp_sum = budget_2020.at[0,'Fund allotted(in crores)']+budget_2020.at[8,'Fund allotted(in crores)']+budget_2020.at[11,'Fund allotted(in crores)']+budget_2020.at[14,'Fund allotted(in crores)']+budget_2020.at[18,'Fund allotted(in crores)']+budget_2020.at[23,'Fund allotted(in crores)']+budget_2020.at[41,'Fund allotted(in crores)']+budget_2020.at[42,'Fund allotted(in crores)']+budget_2020.at[43,'Fund allotted(in crores)']
others = np.sum(budget_2020['Fund allotted(in crores)']) - imp_sum
others
```

Appending 'Others' section to our working dataframe

```python
budget_2020 = budget_2020.iloc[[0,8,11,14,18,23,41,42,43],:]
row = {'Department /Ministry' : 'OTHERS','Fund allotted(in crores)' : others}
budget_2020 = budget_2020.append(row, ignore_index = True)
budget_2020
```

Plotting a bar graph to visually analyze the budget allocated in various departments

```python
budget_2020.plot.bar(x='Department /Ministry',y='Fund allotted(in crores)',title ="Budget Allocation of India - 2020" )
```

Importing the Unemployment rate dataset for 2020 and storing it in a dataframe

```python
unemployment_2020 = pd.read_csv("Unemployment in India - 2020.csv")
unemployment_2020
```

Dropping all the null / empty values

```python
unemployment_2020 = unemployment_2020.dropna()
```

Renaming all the columns because the initial names are not clear

```python
unemployment_2020.columns = ["States","Date","Frequency","Estimated Unemployment Rate","Estimated Employed","Estimated Labour Participation Rate","Region"]
```

Droping 'Date' and 'Frequency' column because it has no role in our analysis

```python
unemployment_2020 = unemployment_2020.drop(['Date','Frequency'],axis=1)
unemployment_2020
```

Plotting a heatmap using seaborn to know the correlation between the features of our dataset

```python
plt.style.use('seaborn-whitegrid')
plt.figure(figsize = (10,8))
plt.title("Correlation between the Features")
sns.heatmap(unemployment_2020.corr())
plt.show
```

Plotting a bar graph to visually analyse the estimated no of employed people in each state of India.

```python
unemployment_2020.columns = ["States","Estimated Unemployment Rate","Estimated Employed","Estimated Labour Participation Rate","Region"]
unemployment_2020.plot.bar(x='States',y='Estimated Employed',title ="Estimated Employment of India - 2021" )
```

Similarly plotting the unemployment rate in each state of India.

```python
unemployment_2020.columns = ["States","Estimated Unemployment Rate","Estimated Employed","Estimated Labour Participation Rate","Region"]
unemployment_2020.plot.bar(x='States',y='Estimated Unemployment Rate',title ="Estimated Unemployment Rate of India - 2020" )
```
